```python
import pandas as pd
```

```python
# Paths
path = '../dataFiles/processedData/countypres_2012_2016_changes.csv'

# Get Data
results_2012_to_2016 = pd.read_csv(path)
```

```python
# Make Election Results FIPS 5 Digits
# Make Sure FIPS are 5 Digits
# Make Int FIPS a String
results_2012_to_2016['FIPS'] = results_2012_to_2016['FIPS'].astype(str)
# If FIPS String is 4 Digits Long- Add 0 to Beginning
results_2012_to_2016.insert(len(results_2012_to_2016.columns), 'Add Zero FIPS', '0' + results_2012_to_2016['FIPS'])
results_2012_to_2016.loc[results_2012_to_2016['FIPS'].str.len()==4, 'FIPS'] = results_2012_to_2016.loc[results_2012_to_2016['FIPS'].str.len()==4, 'Add Zero FIPS']
del results_2012_to_2016['Add Zero FIPS']
results_2012_to_2016
```

```python
# Export Final Dataframe
output_path  = '../dataFiles/mergedData/election_results_2012_to_2016_changes_with_variables.xlsx'
results_2012_to_2016.to_excel(output_path,index=False)
```
