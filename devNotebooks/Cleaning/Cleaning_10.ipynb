```python
import pandas as pd
import numpy as np
```

```python
data2010_to_load = "../Education/ACSDT1Y2010.B07009_data_with_overlays_2019-12-21T130141.csv"
metadata2010_to_load = "../Education/ACSDT1Y2010.B07009_metadata_2019-12-21T130141.csv"

fulldata2010 = pd.read_csv(data2010_to_load)
metadata2010 = pd.read_csv(metadata2010_to_load)
```

```python
fulldata2010.head()
```

```python
data2010 = fulldata2010[['GEO_ID',"NAME","B07009_003E","B07009_004E","B07009_005E","B07009_006E","B07009_001E"]]

data2010 = data2010.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B07009_002E":"Less Than High School Graduate",
                                        "B07009_003E":"High school graduate (includes equivalency)",
                                        "B07009_004E":"Some college or associate's degree",
                                        "B07009_005E":"Bachelor's degree",
                                        "B07009_006E":"Graduate or professional degree","B07009_001E":"Total"})
```

```python
data2010.head()
```

```python
data2010=data2010.drop(index =0)
```

```python
data2010=data2010.reset_index(drop=True)
```

```python
data2010.head()
```

```python
countySplit = data2010["County"].str.split(", ", n = 1, expand = True) 
```

```python
countySplit.head()
```

```python
data2010["County"]= countySplit[0] 
data2010["State"]=countySplit[1]
```

```python
data2010
```

```python
countyDelete = data2010["County"].str.split(" ", n = 1, expand = True) 
```

```python
data2010["County"]= countyDelete[0] 
data2010["Delete"]=countyDelete[1]
```

```python
data2010
```

```python
data2010.drop(columns=['Delete'])
```

```python
splitFIPS = data2010["FIPS"].str.split("US", n = 1, expand = True) 
```

```python
splitFIPS.head()
```

```python
data2010["Delete"]= splitFIPS[0] 
data2010["FIPS"]=splitFIPS[1]
```

```python
data2010.insert(len(data2010.columns), 'Year', 2010)
```

```python
data2010.drop(columns=['Delete'], inplace = True)
```

```python
data2010.head()
```

```python
path = '../dataPoints/education_2010_clean.csv'
data2010.to_csv(path,index=False)
```
