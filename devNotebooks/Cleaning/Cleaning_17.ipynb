```python
data2017_to_load = "../Education/ACSDT1Y2017.B07009_data_with_overlays_2019-12-21T130141.csv"
metadata2017_to_load = "../Education/ACSDT1Y2017.B07009_metadata_2019-12-21T130141.csv"

fulldata2017 = pd.read_csv(data2017_to_load)
metadata2017 = pd.read_csv(metadata2017_to_load)
```
