```python
import pandas as pd
import numpy as np
```

```python
data2012_to_load = "../Education/5Year/ACSDT5Y2012.B06009_data_with_overlays_2020-01-06T222644.csv"
metadata2012_to_load = "../Education/5Year/ACSDT5Y2012.B06009_metadata_2020-01-06T222644.csv"
data2016_to_load = "../Education/5Year/ACSDT5Y2016.B06009_data_with_overlays_2020-01-06T222644.csv"
metadata2016_to_load = "../Education/5Year/ACSDT5Y2016.B06009_metadata_2020-01-06T222644.csv"
data2018_to_load = "../Education/5Year/ACSDT5Y2018.B06009_data_with_overlays_2020-01-06T222644.csv"
metadata2018_to_load = "../Education/5Year/ACSDT5Y2018.B06009_metadata_2020-01-06T222644.csv"

fulldata2012 = pd.read_csv(data2012_to_load)
metadata2012 = pd.read_csv(metadata2012_to_load)
fulldata2016 = pd.read_csv(data2016_to_load)
metadata2016 = pd.read_csv(metadata2016_to_load)
fulldata2018 = pd.read_csv(data2018_to_load)
metadata2018 = pd.read_csv(metadata2018_to_load)
```

```python
data2012 = fulldata2012[['GEO_ID',"NAME","B06009_003E","B06009_004E","B06009_005E","B06009_006E","B06009_001E"]]

data2012 = data2012.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B06009_002E":"Less Than High School Graduate",
                                        "B06009_003E":"High school graduate (includes equivalency)",
                                        "B06009_004E":"Some college or associate's degree",
                                        "B06009_005E":"Bachelor's degree",
                                        "B06009_006E":"Graduate or professional degree","B06009_001E":"Total"})

data2016 = fulldata2016[['GEO_ID',"NAME","B06009_003E","B06009_004E","B06009_005E","B06009_006E","B06009_001E"]]
data2016 = data2016.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B06009_002E":"Less Than High School Graduate",
                                        "B06009_003E":"High school graduate (includes equivalency)",
                                        "B06009_004E":"Some college or associate's degree",
                                        "B06009_005E":"Bachelor's degree",
                                        "B06009_006E":"Graduate or professional degree","B06009_001E":"Total"})

data2018 = fulldata2018[['GEO_ID',"NAME","B06009_003E","B06009_004E","B06009_005E","B06009_006E","B06009_001E"]]

data2018 = data2018.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B06009_002E":"Less Than High School Graduate",
                                        "B06009_003E":"High school graduate (includes equivalency)",
                                        "B06009_004E":"Some college or associate's degree",
                                        "B06009_005E":"Bachelor's degree",
                                        "B06009_006E":"Graduate or professional degree","B06009_001E":"Total"})
```

```python
data2012=data2012.drop(index =0)
data2016=data2016.drop(index =0)
data2018=data2018.drop(index =0)
```
