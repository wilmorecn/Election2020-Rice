```python
import pandas as pd
import numpy as np
```

```python
data2012_to_load = "../Education/ACSDT1Y2012.B07009_data_with_overlays_2019-12-21T130141.csv"
metadata2012_to_load = "../Education/ACSDT1Y2012.B07009_metadata_2019-12-21T130141.csv"

fulldata2012 = pd.read_csv(data2012_to_load)
metadata2012 = pd.read_csv(metadata2012_to_load)
```

```python
fulldata2012.head()
```

```python
data2012 = fulldata2012[['GEO_ID',"NAME","B07009_003E","B07009_004E","B07009_005E","B07009_006E","B07009_001E"]]

data2012 = data2012.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B07009_002E":"Less Than High School Graduate",
                                        "B07009_003E":"High school graduate (includes equivalency)",
                                        "B07009_004E":"Some college or associate's degree",
                                        "B07009_005E":"Bachelor's degree",
                                        "B07009_006E":"Graduate or professional degree","B07009_001E":"Total"})
```

```python
data2012.head()
```

```python
data2012=data2012.drop(index =0)
```

```python
data2012=data2012.reset_index(drop=True)
```

```python
data2012.head()
```

```python
countySplit = data2012["County"].str.split(", ", n = 1, expand = True) 
```

```python
countySplit.head()
```

```python
data2012["County"]= countySplit[0] 
data2012["State"]=countySplit[1]
```

```python
countyDelete = data2012["County"].str.split(" ", n = 1, expand = True) 
```

```python
data2012["County"]= countyDelete[0] 
data2012["Delete"]=countyDelete[1]
```
