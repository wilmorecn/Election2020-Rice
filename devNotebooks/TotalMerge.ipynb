```python
import pandas as pd
import numpy as np
```

```python
year1_to_load = "dataPoints/education_2010-2018.xlsx"
year5_to_load = "dataPoints/education_5_year.xlsx"

year1 = pd.read_excel(year1_to_load)
year5 = pd.read_excel(year5_to_load)
```

```python
year1
```

```python
year1.dropna(inplace = True)
```

```python
complete = year1.merge(year5,how='outer')
```

```python
complete
```

```python
complete = complete[complete.County != 'United']
```

```python
complete
```

```python
complete['FIPS'] = complete['FIPS'].replace('\.0','', regex = True)
complete.head()
```

```python
complete['FIPS'] = complete['FIPS'].astype(str)
```

```python
complete.insert(len(complete.columns), 'Add Zero FIPS', '0' + complete['FIPS'])
complete.loc[complete['FIPS'].str.len()==4, 'FIPS'] = complete.loc[complete['FIPS'].str.len()==4, 'Add Zero FIPS']
del complete['Add Zero FIPS']
```

```python
complete
```

```python
path = '../dataFiles/processedData/education_processed.xlsx'
complete.to_excel(path,index=False)
```
