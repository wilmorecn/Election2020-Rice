```python
import pandas as pd
import numpy as np
```

```python
data2011_to_load = "Education/ACSDT1Y2011.B07009_data_with_overlays_2019-12-21T130141.csv"
metadata2011_to_load = "Education/ACSDT1Y2011.B07009_metadata_2019-12-21T130141.csv"

fulldata2011 = pd.read_csv(data2011_to_load)
metadata2011 = pd.read_csv(metadata2011_to_load)
```

```python
fulldata2011.head()
```

```python
data2011 = fulldata2011[['GEO_ID',"NAME","B07009_003E","B07009_004E","B07009_005E","B07009_006E","B07009_001E"]]

data2011 = data2011.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B07009_002E":"Less Than High School Graduate",
                                        "B07009_003E":"High school graduate (includes equivalency)",
                                        "B07009_004E":"Some college or associate's degree",
                                        "B07009_005E":"Bachelor's degree",
                                        "B07009_006E":"Graduate or professional degree","B07009_001E":"Total"})
```

```python
data2011.head()
```

```python
data2011=data2011.drop(index =0)
```

```python
data2011=data2011.reset_index(drop=True)
```

```python
data2011.head()
```

```python
countySplit = data2011["County"].str.split(", ", n = 1, expand = True) 
```

```python
countySplit.head()
```

```python
data2011["County"]= countySplit[0] 
data2011["State"]=countySplit[1]
```

```python
data2011
```

```python
countyDelete = data2011["County"].str.split(" ", n = 1, expand = True) 
```

```python
data2011["County"]= countyDelete[0] 
data2011["Delete"]=countyDelete[1]
```

```python
data2011.drop(columns=['Delete'])
```

```python
splitFIPS = data2011["FIPS"].str.split("US", n = 1, expand = True) 
```

```python
data2011["Delete"]= splitFIPS[0] 
data2011["FIPS"]=splitFIPS[1]
```
