```python
import pandas as pd
import numpy as np
```

```python
data2015_to_load = "Education/ACSDT1Y2015.B07009_data_with_overlays_2019-12-21T130141.csv"
metadata2015_to_load = "Education/ACSDT1Y2015.B07009_metadata_2019-12-21T130141.csv"

fulldata2015 = pd.read_csv(data2015_to_load)
metadata2015 = pd.read_csv(metadata2015_to_load)
```

```python
data2015 = fulldata2015[['GEO_ID',"NAME","B07009_003E","B07009_004E","B07009_005E","B07009_006E","B07009_001E"]]

data2015 = data2015.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B07009_002E":"Less Than High School Graduate",
                                        "B07009_003E":"High school graduate (includes equivalency)",
                                        "B07009_004E":"Some college or associate's degree",
                                        "B07009_005E":"Bachelor's degree",
                                        "B07009_006E":"Graduate or professional degree","B07009_001E":"Total"})
```

```python
data2015.head()
```

```python
data2015=data2015.drop(index =0)
```

```python
data2015=data2015.reset_index(drop=True)
```

```python
countySplit = data2015["County"].str.split(", ", n = 1, expand = True) 
```

```python
countySplit.head()
```

```python
data2015["County"]= countySplit[0] 
data2015["State"]=countySplit[1]
```
